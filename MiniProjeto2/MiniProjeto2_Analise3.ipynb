```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
df.columns
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=df['fuelType'],
            y=df['price'],
            hue=df['gearbox'],
            palette='Set2')
plt.xlabel('Tipo de Combustível', fontdict={'fontsize':16})
plt.ylabel('Preço', fontdict={'fontsize':16})
plt.title('Preço por tipo de combustível e de câmbio',
            fontdict={'fontsize': 20,
            'fontweight': 'bold'})
plt.savefig('plots/Analise3/preco_tipoCombustívelCambio.png')
plt.show()
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

plt.figure(figsize=(12,8))
sns.barplot(x=df['vehicleType'],
            y=df['powerPS'],
            hue=df['gearbox'],
            palette='Set1')
plt.xlabel('Tipo de Carro', fontdict={'fontsize':16})
plt.ylabel('Potência Média', fontdict={'fontsize':16})
plt.title('Preço por tipo de combustível e de câmbio',
            fontdict={'fontsize': 20,
            'fontweight': 'bold'})
plt.savefig('plots/Analise3/potencia_tipoVeiculo.png')
plt.show()
```
