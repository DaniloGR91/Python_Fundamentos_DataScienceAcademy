## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

```python
df.columns
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

yearReg = df['yearOfRegistration']
plt.figure(figsize=(12,8))
sns.distplot(yearReg, color='green', hist_kws={'histtype': 'stepfilled'})
plt.title('Distribuição dos anos de registro', fontdict={'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('Ano de registro', fontdict={'fontsize': 14})
plt.ylabel('Frequencia', fontdict={'fontsize': 14})
plt.savefig("plots/Analise1/distribuicao_ano.png")
plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers

plt.figure(figsize=(12,8))
sns.boxplot(x = df['vehicleType'], y = df['price'])
plt.xlabel('Tipo de veículo', fontdict={'fontsize': 14})
plt.ylabel('Preço', fontdict={'fontsize':14})
plt.title('Análise de outliers', fontdict={'fontsize': 18, 'fontweight': 'bold', 'color': 'red'})
plt.savefig('plots/Analise1/boxplot_preco.png')
plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
for c, v in enumerate(df['vehicleType'].value_counts()):
    print(c, v)
```

```python
plt.figure(figsize=(12,8))
g = sns.barplot(x=df['vehicleType'].value_counts().keys(), 
                y=df['vehicleType'].value_counts(), 
                palette='Blues_d')
for c, value in enumerate(df['vehicleType'].value_counts()):
    g.text(c-0.2, value+300, value, fontdict={'fontsize': 12, 'fontweight':'bold'})
plt.xlabel('Tipo de Veículo', fontdict={'fontsize': 14})
plt.ylabel('Total', fontdict={'fontsize': 14})
plt.title('Contagem de veículos por tipo', fontdict={'fontsize': 20, 'fontweight': 'bold'})
plt.savefig("plots/Analise1/contagem_tipodeVeiculo.png")
plt.show()
```
