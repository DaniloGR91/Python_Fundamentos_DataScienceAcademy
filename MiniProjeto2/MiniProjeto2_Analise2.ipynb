```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
df.columns
```

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca

plt.figure(figsize=(20,12))
sns.barplot(x=df['brand'].value_counts(), 
            y=df['brand'].value_counts().keys(), 
            orient='h')
plt.xlabel('Número de veículos', fontdict={'fontsize': 16})
plt.ylabel('Marca', fontdict={'fontsize': 16})
plt.title('Veículos por marca', fontdict={'fontsize': 20, 'fontweight': 'bold'})
plt.savefig(("plots/Analise2/contagemPorMarca.png"))
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
df.columns
```

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

plt.figure(figsize=(12,8))
sns.barplot(x=df['vehicleType'], y=df['price'], hue=df['gearbox'])
sns.set_style('whitegrid')
plt.xlabel('Tipo de veículo', fontdict={'fontsize': 16})
plt.ylabel('Preço', fontdict={'fontsize': 16})
plt.title('Preço médio dos veículos por tipo de veículo e câmbio',
            fontdict={'fontsize': 20, 'fontweight':'bold'})
plt.savefig("plots/Analise2/preco_tipoVeiculoCambio.png")
plt.show()
```
