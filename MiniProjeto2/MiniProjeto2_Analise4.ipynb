```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
# Calcule a média de preço por marca e por veículo


df2 = df[['price', 'brand', 'vehicleType']]
df2grouped = df2.groupby(by=['vehicleType', 'brand']).mean()

df3 = df[['brand', 'vehicleType']]
df3 = df3.drop_duplicates()
priceList = []
for car in df3.values:
    priceList.append(df2grouped.loc[car[1],car[0]].values[0])

df3['avgPrice'] = priceList
df3 = df3.round(2)
```

```python
df3.head()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo

palette = sns.cubehelix_palette(10)
fig, ax = plt.subplots(figsize=(16, 30))
df4 = df3.pivot('brand', 'vehicleType', 'avgPrice')
ax = sns.heatmap(df4.fillna(0), 
                cmap='YlGnBu', 
                annot=True, 
                fmt='g', 
                linewidths=.5)
plt.xlabel('Tipo de veículo', 
            fontdict={'fontsize':16})
plt.ylabel('Marca', 
            fontdict={'fontsize': 16})
plt.title('Preço por marca e tipo de veículo', 
            fontdict={'fontsize': 16, 'fontweight': 'bold'})
fig.savefig("plots/Analise4/heatmap_type_price.png")
plt.show()
```
